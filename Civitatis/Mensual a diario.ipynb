```python
#RELLENAR ESTO ANTES DE EJECUTAR

#nombre de la variable de fecha
fecha = 'Fecha'

#nombre de variables a cambiar, y si tiene que ser nº entero o no
# cambiar={'Impresiones':True, 'Clicks':True, 'Inversion':False}
cambiar = {'Imps':True, 'Clicks':True, 'Cost':False}

#aquellas variables que al ser números bajos, no queremos que reparta por toda la columna indiscriminadamente,
#sólo que rellene la cantidad que le corresponde, y luego pare
exacto = ['Sales']

#resto de variables
# resto = ['Anunciante','Hora','Medio', 'Soporte','CampaÃ±a','Producto','Formato']
resto = ['Channel']
```

```python
import datetime #transformo a datetime primero
#.strftime('%d/%m/%Y %H:%M:%S')

df[fecha]=[datetime.datetime.strptime(x, '%d/%m/%Y') for x in df[fecha]]
```

```python
# Get the last day of the month from a given date
import calendar
from datetime import datetime

df_final = df.iloc[[0]] #así tengo un registro para luego poder concatenar con los df. Borro luego

for k in range(len(df)):
    input_dt = df[fecha].iloc[k]
#     print("The original date is:", input_dt.date())

    # monthrange() to gets the date range
    # year = 2022, month = 9
    res = calendar.monthrange(input_dt.year, input_dt.month)
    day = res[1]
#     print(f"Last date of month is: {day}/{input_dt.month}/{input_dt.year}")

    #construyo el df de ese mes
    df_fecha = pd.DataFrame(columns=[fecha],data = pd.date_range(start=df[fecha].iloc[k].strftime('%m/%d/%Y'), 
                  end=str(day)+'/'+str(input_dt.month)+'/'+str(input_dt.year)))

    for i in cambiar: #transformo los numeros
        numero = df[i].iloc[k]
        if cambiar[i]:
            numero_div = round(numero/len(df_fecha),0) #nº entero, sin decimales
        else:
            numero_div = round(numero/len(df_fecha),2) #nº con 2 decimales
        df_fecha[i] = numero_div
        
    for i in exacto:#primero relleno entre todos los días, y luego los sobrantes uno a uno hasta llegar a la cantidad exacta
        numero = df[i].iloc[k]
        df_fecha[i] = numero//len(df_fecha) 
        contador = 0
        faltan = numero - (numero//len(df_fecha))*len(df_fecha) #cantidad que falta por rellenar
        while faltan > contador:
            df_fecha[i].iloc[contador] += 1
            contador +=1
            

    for i in resto: #junto las columnas que no habia que modificar
        df_fecha[i] = df[i].iloc[k]

    #voy juntando los disntpos df
    df_final = pd.concat([df_final,df_fecha], axis=0)

df_final = df_final[1:]
#SI QUIERO FORMATO %d/%m/%Y %H:%M:%S, DESCOMENTAR EL DE ABAJO Y COMENTAR EL SIGUIENTE
# df_final[fecha] = [x.strftime('%d/%m/%Y %H:%M:%S') for x in df_final[fecha]]
df_final[fecha] = [x.strftime('%d/%m/%Y') for x in df_final[fecha]]
df_final
```
